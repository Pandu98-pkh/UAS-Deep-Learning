## 1. Apa itu Machine Learning?

### Definisi dan Konsep Dasar

Machine Learning adalah ilmu (dan seni) memprogram komputer agar dapat belajar dari data, tanpa harus diprogram secara eksplisit untuk setiap kasus.

**Definisi Formal menurut Tom Mitchell (1997)**: 
"Sebuah program komputer dikatakan belajar dari pengalaman E dalam menyelesaikan tugas T dan diukur dengan performa P, jika performanya pada T, yang diukur dengan P, meningkat dengan pengalaman E."

## 2. Jenis-jenis Sistem Machine Learning

### 2.1 Berdasarkan Supervisi: Supervised vs Unsupervised Learning

```python
# Demonstrasi perbedaan Supervised vs Unsupervised Learning
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_classification, make_blobs
from sklearn.cluster import KMeans

# Generate data untuk supervised learning (classification)
X_supervised, y_supervised = make_classification(n_samples=100, n_features=2, 
                                               n_redundant=0, n_informative=2,
                                               random_state=42, n_clusters_per_class=1)

# Generate data untuk unsupervised learning (clustering)
X_unsupervised, _ = make_blobs(n_samples=100, centers=3, n_features=2, 
                              random_state=42, cluster_std=1.5)

# Visualisasi
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 4))

# Supervised Learning - dengan label
scatter1 = ax1.scatter(X_supervised[:, 0], X_supervised[:, 1], c=y_supervised, 
                      cmap='viridis', alpha=0.7)
ax1.set_title('Supervised Learning\n(Data dengan Label)', fontweight='bold')
ax1.set_xlabel('Feature 1')
ax1.set_ylabel('Feature 2')
plt.colorbar(scatter1, ax=ax1)

# Unsupervised Learning - tanpa label
ax2.scatter(X_unsupervised[:, 0], X_unsupervised[:, 1], alpha=0.7, color='gray')
ax2.set_title('Unsupervised Learning\n(Data tanpa Label)', fontweight='bold')
ax2.set_xlabel('Feature 1')
ax2.set_ylabel('Feature 2')

# Unsupervised Learning - hasil clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_unsupervised)
scatter3 = ax3.scatter(X_unsupervised[:, 0], X_unsupervised[:, 1], c=clusters, 
                      cmap='viridis', alpha=0.7)
ax3.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
           marker='x', s=200, linewidths=3, color='red')
ax3.set_title('Unsupervised Learning\n(Hasil Clustering)', fontweight='bold')
ax3.set_xlabel('Feature 1')
ax3.set_ylabel('Feature 2')
plt.colorbar(scatter3, ax=ax3)

plt.tight_layout()
plt.show()

print("=== JENIS-JENIS SUPERVISED LEARNING ===")
print("1. CLASSIFICATION (Klasifikasi)")
print("   - Tujuan: Memprediksi kategori/kelas")
print("   - Contoh: Spam detection, image recognition")
print("   - Output: Diskrit (kategori)")

print("\n2. REGRESSION (Regresi)")
print("   - Tujuan: Memprediksi nilai numerik")
print("   - Contoh: Prediksi harga rumah, suhu")
print("   - Output: Kontinu (angka)")

print("\n=== JENIS-JENIS UNSUPERVISED LEARNING ===")
print("1. CLUSTERING")
print("   - Tujuan: Mengelompokkan data serupa")
print("   - Contoh: Segmentasi pelanggan")

print("2. DIMENSIONALITY REDUCTION")
print("   - Tujuan: Mengurangi fitur sambil mempertahankan informasi")
print("   - Contoh: Visualisasi data high-dimensional")

print("3. ANOMALY DETECTION")
print("   - Tujuan: Mendeteksi data yang tidak normal")
print("   - Contoh: Deteksi fraud")
```

**Penjelasan Teoretis:**

**Supervised Learning:**
- **Mengapa disebut "supervised"?** Karena algoritma belajar dari data yang sudah berlabel, seperti guru yang memberikan jawaban yang benar
- **Bagaimana cara kerjanya?** Algoritma mencari pola hubungan antara input (fitur) dan output (label) dari data training, kemudian menggunakan pola tersebut untuk prediksi data baru
- **Kapan digunakan?** Ketika kita punya data historis dengan jawaban yang sudah diketahui dan ingin memprediksi jawaban untuk data baru

**Unsupervised Learning:**
- **Mengapa disebut "unsupervised"?** Karena tidak ada "guru" atau label yang memberitahu jawaban yang benar
- **Bagaimana cara kerjanya?** Algoritma mencari pola tersembunyi atau struktur dalam data tanpa panduan eksternal
- **Kapan digunakan?** Ketika ingin menemukan pola tersembunyi dalam data atau tidak memiliki data berlabel

### 2.2 Berdasarkan Metode Pembelajaran: Batch vs Online Learning

```python
# Simulasi Batch Learning vs Online Learning
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler

# Generate synthetic streaming data
np.random.seed(42)
n_samples = 1000
X_stream = np.random.randn(n_samples, 1)
y_stream = 2 * X_stream.ravel() + 1 + 0.1 * np.random.randn(n_samples)

# Batch Learning - menggunakan semua data sekaligus
print("=== BATCH LEARNING ===")
print("Karakteristik:")
print("- Menggunakan semua data training sekaligus")
print("- Proses training membutuhkan waktu lama")
print("- Setelah training, model tidak bisa belajar lagi")
print("- Cocok untuk data yang tidak berubah")

# Online Learning - belajar secara bertahap
print("\n=== ONLINE LEARNING ===")
print("Karakteristik:")
print("- Belajar dari data satu per satu atau dalam batch kecil")
print("- Dapat beradaptasi dengan data baru")
print("- Cocok untuk data streaming atau big data")
print("- Learning rate mengontrol seberapa cepat model beradaptasi")

# Simulasi Online Learning dengan berbagai learning rate
learning_rates = [0.01, 0.1, 0.3]
colors = ['blue', 'green', 'red']

plt.figure(figsize=(15, 5))

for i, (lr, color) in enumerate(zip(learning_rates, colors)):
    plt.subplot(1, 3, i+1)
    
    # Online SGD model
    model = SGDRegressor(learning_rate='constant', eta0=lr, random_state=42)
    
    # Simulate online learning
    predictions = []
    batch_size = 10
    
    for start in range(0, min(200, len(X_stream)), batch_size):
        end = min(start + batch_size, len(X_stream))
        X_batch = X_stream[start:end]
        y_batch = y_stream[start:end]
        
        if start == 0:
            model.fit(X_batch, y_batch)
        else:
            model.partial_fit(X_batch, y_batch)
        
        # Make prediction on test point
        pred = model.predict([[0]])[0]
        predictions.append(pred)
    
    # Plot convergence
    plt.plot(predictions, color=color, linewidth=2)
    plt.axhline(y=1, color='black', linestyle='--', alpha=0.5, label='True value')
    plt.title(f'Learning Rate = {lr}', fontweight='bold')
    plt.xlabel('Update Steps')
    plt.ylabel('Prediction')
    plt.grid(True, alpha=0.3)
    if i == 0:
        plt.legend()

plt.tight_layout()
plt.show()

print("\n=== PENGARUH LEARNING RATE ===")
print("- Learning Rate Tinggi: Model belajar cepat tapi bisa unstable")
print("- Learning Rate Rendah: Model belajar stabil tapi lambat")
print("- Learning Rate Optimal: Balance antara kecepatan dan stabilitas")
```

**Penjelasan Teoretis:**

**Batch Learning:**
- **Mengapa disebut "batch"?** Karena semua data diproses dalam satu batch besar sekaligus
- **Bagaimana cara kerjanya?** Model dilatih sekali menggunakan seluruh dataset, setelah itu model tidak bisa diupdate lagi tanpa retraining
- **Keuntungan:** Sederhana, stabil, cocok untuk dataset kecil-menengah
- **Kekurangan:** Tidak bisa beradaptasi dengan data baru, membutuhkan resource besar

**Online Learning:**
- **Mengapa disebut "online"?** Karena model dapat belajar secara "live" dari data yang datang secara bertahap
- **Bagaimana cara kerjanya?** Model diupdate secara incremental setiap kali menerima data baru
- **Parameter penting:** Learning rate mengontrol seberapa cepat model beradaptasi
- **Keuntungan:** Adaptif, cocok untuk big data dan data streaming
- **Kekurangan:** Lebih kompleks, rentan terhadap data buruk

### 2.3 Berdasarkan Metode Generalisasi: Instance-based vs Model-based Learning

```python
# Demonstrasi Instance-based vs Model-based Learning
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data - hubungan GDP per capita dengan life satisfaction
np.random.seed(42)
gdp_per_capita = np.random.uniform(10000, 60000, 20)
life_satisfaction = 4.5 + (gdp_per_capita - 10000) * 2.5e-5 + np.random.normal(0, 0.5, 20)

# Prepare data
X = gdp_per_capita.reshape(-1, 1)
y = life_satisfaction

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Instance-based Learning (K-Nearest Neighbors)
knn_model = KNeighborsRegressor(n_neighbors=3)
knn_model.fit(X_train, y_train)

# Model-based Learning (Linear Regression)
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

# Create prediction range
X_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
knn_pred = knn_model.predict(X_range)
linear_pred = linear_model.predict(X_range)

# Plot comparison
plt.figure(figsize=(15, 5))

# Instance-based (KNN)
plt.subplot(1, 3, 1)
plt.scatter(X_train, y_train, color='blue', alpha=0.7, label='Training Data')
plt.scatter(X_test, y_test, color='red', alpha=0.7, label='Test Data')
plt.plot(X_range, knn_pred, color='green', linewidth=2, label='KNN Prediction')
plt.title('Instance-based Learning\n(K-Nearest Neighbors)', fontweight='bold')
plt.xlabel('GDP per Capita')
plt.ylabel('Life Satisfaction')
plt.legend()
plt.grid(True, alpha=0.3)

# Model-based (Linear Regression)
plt.subplot(1, 3, 2)
plt.scatter(X_train, y_train, color='blue', alpha=0.7, label='Training Data')
plt.scatter(X_test, y_test, color='red', alpha=0.7, label='Test Data')
plt.plot(X_range, linear_pred, color='orange', linewidth=2, label='Linear Model')
plt.title('Model-based Learning\n(Linear Regression)', fontweight='bold')
plt.xlabel('GDP per Capita')
plt.ylabel('Life Satisfaction')
plt.legend()
plt.grid(True, alpha=0.3)

# Comparison
plt.subplot(1, 3, 3)
plt.scatter(X_train, y_train, color='blue', alpha=0.7, label='Training Data')
plt.plot(X_range, knn_pred, color='green', linewidth=2, label='KNN', linestyle='--')
plt.plot(X_range, linear_pred, color='orange', linewidth=2, label='Linear', linestyle='-')
plt.title('Perbandingan Kedua Metode', fontweight='bold')
plt.xlabel('GDP per Capita')
plt.ylabel('Life Satisfaction')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Performance comparison
knn_score = knn_model.score(X_test, y_test)
linear_score = linear_model.score(X_test, y_test)

print("=== PERBANDINGAN PERFORMANCE ===")
print(f"Instance-based (KNN) R² Score: {knn_score:.3f}")
print(f"Model-based (Linear) R² Score: {linear_score:.3f}")

print("\n=== KARAKTERISTIK INSTANCE-BASED LEARNING ===")
print("- Menyimpan semua data training")
print("- Prediksi berdasarkan similaritas dengan data training")
print("- Tidak ada fase training eksplisit")
print("- Prediksi lambat, storage besar")
print("- Contoh: K-Nearest Neighbors, Case-based Reasoning")

print("\n=== KARAKTERISTIK MODEL-BASED LEARNING ===")
print("- Membangun model matematis dari data training")
print("- Prediksi berdasarkan parameter model")
print("- Ada fase training untuk optimasi parameter")
print("- Prediksi cepat, storage kecil")
print("- Contoh: Linear Regression, Neural Networks")
```

**Penjelasan Teoretis:**

**Instance-based Learning:**
- **Mengapa disebut "instance-based"?** Karena prediksi dibuat berdasarkan kemiripan dengan instance (contoh) yang ada di data training
- **Bagaimana cara kerjanya?** Menyimpan semua data training, saat prediksi mencari data training yang paling mirip
- **Keuntungan:** Sederhana, tidak perlu asumsi tentang bentuk data, baik untuk data lokal
- **Kekurangan:** Membutuhkan storage besar, prediksi lambat, sensitif terhadap noise

**Model-based Learning:**
- **Mengapa disebut "model-based"?** Karena membuat model matematis yang menggeneralisasi pola dalam data
- **Bagaimana cara kerjanya?** Belajar parameter model dari data training, menggunakan model untuk prediksi
- **Keuntungan:** Prediksi cepat, storage efisien, dapat mengekstrapolasi
- **Kekurangan:** Perlu asumsi tentang bentuk model, mungkin terlalu simpel untuk data kompleks

## 3. Implementasi Praktis: Contoh Linear Regression

Sekarang mari kita implementasikan contoh yang ada di buku - prediksi life satisfaction berdasarkan GDP per capita:

```python
# Reproduksi kode dari buku - Example 1-1
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.linear_model
import sklearn.neighbors

# Fungsi untuk simulasi data (karena kita tidak punya file asli)
def prepare_country_stats():
    """
    Simulasi fungsi prepare_country_stats dari buku
    Mengembalikan data GDP per capita dan Life satisfaction
    """
    # Data simulasi berdasarkan contoh di buku
    data = {
        'Country': ['Hungary', 'Korea', 'France', 'Australia', 'United States'],
        'GDP per capita': [12240, 27195, 37675, 50962, 55805],
        'Life satisfaction': [4.9, 5.8, 6.5, 7.3, 7.2]
    }
    return pd.DataFrame(data)

# Load dan prepare data
print("=== LOADING DATA ===")
country_stats = prepare_country_stats()
print(country_stats)

# Prepare features dan target
X = country_stats[["GDP per capita"]].values
y = country_stats[["Life satisfaction"]].values

print(f"\nShape of X (features): {X.shape}")
print(f"Shape of y (target): {y.shape}")

# Visualisasi data
plt.figure(figsize=(10, 6))
plt.scatter(country_stats["GDP per capita"], country_stats["Life satisfaction"], 
           alpha=0.8, s=100)

# Tambahkan label untuk setiap negara
for i, country in enumerate(country_stats['Country']):
    plt.annotate(country, 
                (country_stats["GDP per capita"].iloc[i], 
                 country_stats["Life satisfaction"].iloc[i]),
                xytext=(5, 5), textcoords='offset points', fontsize=9)

plt.xlabel('GDP per Capita (USD)')
plt.ylabel('Life Satisfaction')
plt.title('Life Satisfaction vs GDP per Capita', fontweight='bold')
plt.grid(True, alpha=0.3)
plt.show()

# Select dan train linear model
print("\n=== TRAINING LINEAR REGRESSION MODEL ===")
model = sklearn.linear_model.LinearRegression()

# Train the model
model.fit(X, y)

print(f"Model coefficients (slope): {model.coef_[0][0]:.2e}")
print(f"Model intercept: {model.intercept_[0]:.2f}")

# Visualisasi model
plt.figure(figsize=(10, 6))

# Plot data points
plt.scatter(X, y, alpha=0.8, s=100, label='Data')

# Plot regression line
X_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_pred_range = model.predict(X_range)
plt.plot(X_range, y_pred_range, 'r-', linewidth=2, label='Linear Model')

# Equation
equation = f'Life Satisfaction = {model.intercept_[0]:.2f} + {model.coef_[0][0]:.2e} × GDP'
plt.text(0.05, 0.95, equation, transform=plt.gca().transAxes, 
         bbox=dict(boxstyle="round", facecolor='wheat', alpha=0.8),
         verticalalignment='top', fontsize=10)

plt.xlabel('GDP per Capita (USD)')
plt.ylabel('Life Satisfaction')
plt.title('Linear Regression Model', fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Make prediction untuk Cyprus (contoh dari buku)
print("\n=== MAKING PREDICTIONS ===")
cyprus_gdp = np.array([[22587]])  # Cyprus's GDP per capita
cyprus_prediction = model.predict(cyprus_gdp)
print(f"Cyprus GDP per capita: ${cyprus_gdp[0][0]:,}")
print(f"Predicted life satisfaction: {cyprus_prediction[0][0]:.2f}")

# Bandingkan dengan K-Nearest Neighbors
print("\n=== COMPARISON WITH K-NEAREST NEIGHBORS ===")
knn_model = sklearn.neighbors.KNeighborsRegressor(n_neighbors=3)
knn_model.fit(X, y)
cyprus_knn_prediction = knn_model.predict(cyprus_gdp)

print(f"Linear Regression prediction: {cyprus_prediction[0][0]:.2f}")
print(f"K-NN prediction: {cyprus_knn_prediction[0][0]:.2f}")

# Visualisasi perbandingan
plt.figure(figsize=(12, 5))

# Linear Regression
plt.subplot(1, 2, 1)
plt.scatter(X, y, alpha=0.8, s=100, label='Training Data')
plt.plot(X_range, y_pred_range, 'r-', linewidth=2, label='Linear Model')
plt.scatter(cyprus_gdp, cyprus_prediction, color='red', s=200, marker='*', 
           label=f'Cyprus Prediction: {cyprus_prediction[0][0]:.2f}')
plt.xlabel('GDP per Capita (USD)')
plt.ylabel('Life Satisfaction')
plt.title('Linear Regression', fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)

# K-Nearest Neighbors
plt.subplot(1, 2, 2)
plt.scatter(X, y, alpha=0.8, s=100, label='Training Data')
knn_pred_range = knn_model.predict(X_range)
plt.plot(X_range, knn_pred_range, 'g-', linewidth=2, label='KNN Model')
plt.scatter(cyprus_gdp, cyprus_knn_prediction, color='green', s=200, marker='*', 
           label=f'Cyprus Prediction: {cyprus_knn_prediction[0][0]:.2f}')
plt.xlabel('GDP per Capita (USD)')
plt.ylabel('Life Satisfaction')
plt.title('K-Nearest Neighbors', fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

**Penjelasan Teoretis Implementasi:**

**Linear Regression:**
- **Bagaimana cara kerjanya?** Mencari garis lurus terbaik yang meminimalkan jarak antara prediksi dan data aktual
- **Parameter yang dipelajari:** Slope (kemiringan) dan intercept (titik potong dengan sumbu y)
- **Fungsi cost:** Mean Squared Error - mengukur rata-rata kuadrat selisih prediksi dengan aktual
- **Optimasi:** Menggunakan Normal Equation atau Gradient Descent untuk menemukan parameter optimal

**Proses Training:**
1. **Data Preparation:** Menyiapkan fitur (X) dan target (y)
2. **Model Selection:** Memilih Linear Regression sebagai model
3. **Training:** Algoritma mencari parameter terbaik menggunakan data training
4. **Prediction:** Menggunakan model terlatih untuk prediksi data baru

## 4. Tantangan Utama dalam Machine Learning

```python
# Demonstrasi pengaruh jumlah data training terhadap performa model
from sklearn.model_selection import learning_curve
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
n_samples = 500
X_full = np.random.uniform(0, 10, n_samples).reshape(-1, 1)
y_full = 2 * X_full.ravel() + 1 + np.sin(X_full.ravel()) + np.random.normal(0, 0.5, n_samples)

# Create models with different complexity
models = {
    'Linear': LinearRegression(),
    'Polynomial (degree=3)': Pipeline([
        ('poly', PolynomialFeatures(degree=3)),
        ('linear', LinearRegression())
    ])
}

plt.figure(figsize=(15, 10))

for i, (name, model) in enumerate(models.items()):
    # Learning curve
    train_sizes, train_scores, val_scores = learning_curve(
        model, X_full, y_full, 
        train_sizes=np.linspace(0.1, 1.0, 10),
        cv=5, scoring='neg_mean_squared_error'
    )
    
    train_mean = -train_scores.mean(axis=1)
    train_std = train_scores.std(axis=1)
    val_mean = -val_scores.mean(axis=1)
    val_std = val_scores.std(axis=1)
    
    plt.subplot(2, 2, i+1)
    plt.plot(train_sizes, train_mean, 'o-', color='blue', label='Training Error')
    plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1, color='blue')
    plt.plot(train_sizes, val_mean, 'o-', color='red', label='Validation Error')
    plt.fill_between(train_sizes, val_mean - val_std, val_mean + val_std, alpha=0.1, color='red')
    plt.xlabel('Training Set Size')
    plt.ylabel('Mean Squared Error')
    plt.title(f'{name}', fontweight='bold')
    plt.legend()
    plt.grid(True, alpha=0.3)

# Visualisasi data yang terbatas vs banyak
plt.subplot(2, 2, 3)
# Few data points
few_indices = np.random.choice(len(X_full), 20, replace=False)
X_few = X_full[few_indices]
y_few = y_full[few_indices]

model_few = LinearRegression()
model_few.fit(X_few, y_few)

X_plot = np.linspace(0, 10, 100).reshape(-1, 1)
y_plot_few = model_few.predict(X_plot)

plt.scatter(X_few, y_few, alpha=0.7, color='red', s=50)
plt.plot(X_plot, y_plot_few, color='red', linewidth=2, label='Model (20 samples)')
plt.title('Model dengan Data Sedikit', fontweight='bold')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(2, 2, 4)
# Many data points
model_many = LinearRegression()
model_many.fit(X_full, y_full)
y_plot_many = model_many.predict(X_plot)

plt.scatter(X_full, y_full, alpha=0.3, color='blue', s=10)
plt.plot(X_plot, y_plot_many, color='blue', linewidth=2, label='Model (500 samples)')
plt.title('Model dengan Data Banyak', fontweight='bold')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("=== THE UNREASONABLE EFFECTIVENESS OF DATA ===")
print("Penelitian menunjukkan bahwa:")
print("1. Algoritma sederhana dengan data banyak > Algoritma kompleks dengan data sedikit")
print("2. Performa meningkat logaritmik terhadap jumlah data")
print("3. Big data sering lebih penting daripada algoritma canggih")
print("4. Tapi small/medium dataset masih umum - jangan abaikan algoritma!")
```

### 4.3 Poor Quality Data

```python
# Demonstrasi pengaruh kualitas data terhadap model
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Generate clean data
np.random.seed(42)
X_clean = np.random.uniform(0, 10, 100).reshape(-1, 1)
y_clean = 2 * X_clean.ravel() + 1 + np.random.normal(0, 0.2, 100)

# Add various types of noise/errors
def add_outliers(X, y, n_outliers=5):
    """Add outliers to data"""
    X_noisy = X.copy()
    y_noisy = y.copy()
    
    outlier_indices = np.random.choice(len(X), n_outliers, replace=False)
    y_noisy[outlier_indices] += np.random.normal(0, 5, n_outliers)
    
    return X_noisy, y_noisy

def add_missing_values(X, y, missing_rate=0.1):
    """Simulate missing values"""
    X_missing = X.copy()
    y_missing = y.copy()
    
    n_missing = int(len(X) * missing_rate)
    missing_indices = np.random.choice(len(X), n_missing, replace=False)
    
    # Set missing values to NaN
    X_missing[missing_indices] = np.nan
    
    return X_missing, y_missing

def add_noise(X, y, noise_level=1.0):
    """Add random noise to data"""
    X_noisy = X + np.random.normal(0, noise_level * 0.1, X.shape)
    y_noisy = y + np.random.normal(0, noise_level, y.shape)
    return X_noisy, y_noisy

# Create different quality datasets
X_outliers, y_outliers = add_outliers(X_clean, y_clean)
X_missing, y_missing = add_missing_values(X_clean, y_clean)
X_noisy, y_noisy = add_noise(X_clean, y_clean, noise_level=2.0)

# Handle missing values (simple imputation)
X_missing_clean = X_missing.copy()
y_missing_clean = y_missing.copy()
missing_mask = ~np.isnan(X_missing_clean).ravel()
X_missing_clean = X_missing_clean[missing_mask]
y_missing_clean = y_missing_clean[missing_mask]

# Train models
models = {}
datasets = {
    'Clean Data': (X_clean, y_clean),
    'With Outliers': (X_outliers, y_outliers),
    'With Missing Values': (X_missing_clean, y_missing_clean),
    'Noisy Data': (X_noisy, y_noisy)
}

plt.figure(figsize=(16, 10))

for i, (name, (X, y)) in enumerate(datasets.items()):
    # Train model
    model = LinearRegression()
    model.fit(X, y)
    models[name] = model
    
    # Plot
    plt.subplot(2, 2, i+1)
    plt.scatter(X, y, alpha=0.6, s=30)
    
    # Prediction line
    X_range = np.linspace(X_clean.min(), X_clean.max(), 100).reshape(-1, 1)
    y_pred = model.predict(X_range)
    plt.plot(X_range, y_pred, 'r-', linewidth=2, label=f'Model')
    
    # True relationship (from clean data)
    true_model = LinearRegression()
    true_model.fit(X_clean, y_clean)
    y_true = true_model.predict(X_range)
    plt.plot(X_range, y_true, 'g--', linewidth=2, alpha=0.7, label='True Relationship')
    
    plt.title(name, fontweight='bold')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Compare model performance
print("=== PERBANDINGAN PERFORMANCE MODEL ===")
X_test_range = np.linspace(0, 10, 50).reshape(-1, 1)
y_true_test = true_model.predict(X_test_range)

for name, model in models.items():
    y_pred_test = model.predict(X_test_range)
    mse = mean_squared_error(y_true_test, y_pred_test)
    print(f"{name:20}: MSE = {mse:.3f}")

print("\n=== STRATEGI DATA CLEANING ===")
print("\n1. HANDLING OUTLIERS:")
print("   - Deteksi: Z-score, IQR method, Isolation Forest")
print("   - Treatment: Remove, transform, atau robust models")

print("\n2. HANDLING MISSING VALUES:")
print("   - Simple: Remove rows/columns dengan missing values")
print("   - Imputation: Mean, median, mode, atau advanced methods")
print("   - Model-based: Predict missing values")

print("\n3. NOISE REDUCTION:")
print("   - Data validation dan quality checks")
print("   - Smoothing techniques")
print("   - Robust algorithms yang tahan noise")
```

### 4.4 Overfitting vs Underfitting

```python
# Demonstrasi Overfitting dan Underfitting
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.model_selection import validation_curve
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data dengan noise
np.random.seed(42)
n_samples = 100
X_demo = np.random.uniform(0, 1, n_samples).reshape(-1, 1)
y_demo = 1.5 * X_demo.ravel() + 0.5 * np.sin(2 * np.pi * X_demo.ravel()) + np.random.normal(0, 0.1, n_samples)

# Test different polynomial degrees
degrees = range(1, 16)
train_scores, val_scores = validation_curve(
    Pipeline([('poly', PolynomialFeatures()), ('linear', LinearRegression())]),
    X_demo, y_demo, param_name='poly__degree', param_range=degrees,
    cv=5, scoring='neg_mean_squared_error'
)

# Plot bias-variance tradeoff
plt.figure(figsize=(15, 10))

# Validation curve
plt.subplot(2, 3, 1)
train_mean = -train_scores.mean(axis=1)
train_std = train_scores.std(axis=1)
val_mean = -val_scores.mean(axis=1)
val_std = val_scores.std(axis=1)

plt.plot(degrees, train_mean, 'o-', color='blue', label='Training Error')
plt.fill_between(degrees, train_mean - train_std, train_mean + train_std, alpha=0.1, color='blue')
plt.plot(degrees, val_mean, 'o-', color='red', label='Validation Error')
plt.fill_between(degrees, val_mean - val_std, val_mean + val_std, alpha=0.1, color='red')
plt.xlabel('Polynomial Degree')
plt.ylabel('Mean Squared Error')
plt.title('Bias-Variance Tradeoff', fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)

# Show examples of underfitting, good fit, and overfitting
X_plot = np.linspace(0, 1, 100).reshape(-1, 1)
example_degrees = [1, 4, 15]
titles = ['Underfitting\n(High Bias)', 'Good Fit\n(Balanced)', 'Overfitting\n(High Variance)']

for i, (degree, title) in enumerate(zip(example_degrees, titles)):
    plt.subplot(2, 3, i+2)
    
    # Fit model
    poly_model = Pipeline([
        ('poly', PolynomialFeatures(degree=degree)),
        ('linear', LinearRegression())
    ])
    poly_model.fit(X_demo, y_demo)
    y_plot = poly_model.predict(X_plot)
    
    # Plot
    plt.scatter(X_demo, y_demo, alpha=0.6, s=30, color='blue', label='Data')
    plt.plot(X_plot, y_plot, 'r-', linewidth=2, label=f'Degree {degree}')
    plt.title(title, fontweight='bold')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.ylim(-1, 3)

# Regularization demonstration
plt.subplot(2, 3, 5)
from sklearn.linear_model import Ridge

# High degree with regularization
degrees_reg = [15]
alphas = [0, 0.1, 1, 10]
colors = ['red', 'green', 'blue', 'orange']

for alpha, color in zip(alphas, colors):
    ridge_model = Pipeline([
        ('poly', PolynomialFeatures(degree=15)),
        ('ridge', Ridge(alpha=alpha))
    ])
    ridge_model.fit(X_demo, y_demo)
    y_plot_ridge = ridge_model.predict(X_plot)
    
    plt.plot(X_plot, y_plot_ridge, color=color, linewidth=2, 
             label=f'α = {alpha}')

plt.scatter(X_demo, y_demo, alpha=0.6, s=30, color='black', label='Data')
plt.title('Regularization Effect\n(Degree 15)', fontweight='bold')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True, alpha=0.3)
plt.ylim(-1, 3)

# Learning curves to show overfitting
plt.subplot(2, 3, 6)
from sklearn.model_selection import learning_curve

# Complex model (degree 15)
train_sizes, train_scores_lc, val_scores_lc = learning_curve(
    Pipeline([('poly', PolynomialFeatures(degree=15)), ('linear', LinearRegression())]),
    X_demo, y_demo, train_sizes=np.linspace(0.1, 1.0, 10), cv=5, scoring='neg_mean_squared_error'
)

plt.plot(train_sizes, -train_scores_lc.mean(axis=1), 'o-', color='blue', label='Training Error')
plt.plot(train_sizes, -val_scores_lc.mean(axis=1), 'o-', color='red', label='Validation Error')
plt.xlabel('Training Set Size')
plt.ylabel('Mean Squared Error')
plt.title('Learning Curve\n(Overfitting Model)', fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("=== OVERFITTING ===")
print("Ciri-ciri:")
print("- Training error rendah, validation error tinggi")
print("- Model terlalu kompleks untuk data yang ada")
print("- Performance gap besar antara training dan test")

print("\nSolusi:")
print("- Simplify model (reduce parameters)")
print("- Regularization (L1, L2)")
print("- More training data")
print("- Cross-validation")

print("\n=== UNDERFITTING ===")
print("Ciri-ciri:")
print("- Training error dan validation error sama-sama tinggi")
print("- Model terlalu simpel untuk menangkap pola")
print("- High bias")

print("\nSolusi:")
print("- Use more complex model")
print("- Add more features")
print("- Reduce regularization")
print("- Feature engineering")

print("\n=== REGULARIZATION ===")
print("Konsep:")
print("- Menambah penalty term pada cost function")
print("- Mendorong model untuk lebih simple")
print("- L1 (Lasso): Sparsity, feature selection")
print("- L2 (Ridge): Smooth coefficients")
print("- Hyperparameter α mengontrol strength")
```

**Penjelasan Teoretis Overfitting dan Underfitting:**

**Overfitting:**
- **Mengapa terjadi?** Model terlalu kompleks sehingga "menghafal" noise dalam data training
- **Bagaimana mendeteksi?** Training error rendah tapi validation error tinggi
- **Dampak:** Model tidak bisa generalize ke data baru
- **Analogi:** Seperti siswa yang menghafal soal ujian tapi tidak memahami konsepnya

**Underfitting:**
- **Mengapa terjadi?** Model terlalu sederhana untuk menangkap pola yang kompleks
- **Bagaimana mendeteksi?** Training error dan validation error sama-sama tinggi
- **Dampak:** Model tidak bisa belajar dari data dengan baik
- **Analogi:** Seperti menggunakan penggaris untuk mengukur benda yang melingkar

**Regularization:**
- **Tujuan:** Mencegah overfitting dengan menambah constraint pada model
- **Cara kerja:** Menambah penalty term yang menghukum parameter yang terlalu besar
- **Trade-off:** Sedikit mengurangi performa pada training data untuk meningkatkan generalization

## 5. Testing dan Validation

### 5.1 Train-Test Split

```python
# Demonstrasi pentingnya train-test split
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
n_samples = 200
X_full = np.random.uniform(0, 10, n_samples).reshape(-1, 1)
y_full = 2 * X_full.ravel() + 1 + 0.3 * np.sin(2 * X_full.ravel()) + np.random.normal(0, 0.5, n_samples)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_full, y_full, test_size=0.2, random_state=42)

print("=== DATA SPLITTING ===")
print(f"Total samples: {len(X_full)}")
print(f"Training samples: {len(X_train)} ({len(X_train)/len(X_full)*100:.1f}%)")
print(f"Test samples: {len(X_test)} ({len(X_test)/len(X_full)*100:.1f}%)")

# Train models dengan kompleksitas berbeda
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline

models = {
    'Linear': Pipeline([('poly', PolynomialFeatures(1)), ('linear', LinearRegression())]),
    'Polynomial (degree=3)': Pipeline([('poly', PolynomialFeatures(3)), ('linear', LinearRegression())]),
    'Polynomial (degree=10)': Pipeline([('poly', PolynomialFeatures(10)), ('linear', LinearRegression())])
}

results = {}
plt.figure(figsize=(15, 10))

for i, (name, model) in enumerate(models.items()):
    # Train model
    model.fit(X_train, y_train)
    
    # Predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    # Calculate metrics
    train_mse = mean_squared_error(y_train, y_train_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)
    train_r2 = r2_score(y_train, y_train_pred)
    test_r2 = r2_score(y_test, y_test_pred)
    
    results[name] = {
        'train_mse': train_mse, 'test_mse': test_mse,
        'train_r2': train_r2, 'test_r2': test_r2
    }
    
    # Plot
    plt.subplot(2, 3, i+1)
    
    # Plot training data
    plt.scatter(X_train, y_train, alpha=0.6, s=30, color='blue', label='Training')
    plt.scatter(X_test, y_test, alpha=0.6, s=30, color='red', label='Test')
    
    # Plot prediction line
    X_plot = np.linspace(0, 10, 100).reshape(-1, 1)
    y_plot = model.predict(X_plot)
    plt.plot(X_plot, y_plot, 'g-', linewidth=2, label='Model')
    
    plt.title(f'{name}\nTrain R²: {train_r2:.3f}, Test R²: {test_r2:.3f}', fontweight='bold')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True, alpha=0.3)

# Performance comparison
plt.subplot(2, 3, 4)
model_names = list(results.keys())
train_mses = [results[name]['train_mse'] for name in model_names]
test_mses = [results[name]['test_mse'] for name in model_names]

x_pos = np.arange(len(model_names))
width = 0.35

plt.bar(x_pos - width/2, train_mses, width, label='Training MSE', alpha=0.7)
plt.bar(x_pos + width/2, test_mses, width, label='Test MSE', alpha=0.7)
plt.xlabel('Models')
plt.ylabel('Mean Squared Error')
plt.title('Training vs Test Performance', fontweight='bold')
plt.xticks(x_pos, [name.split('(')[0].strip() for name in model_names], rotation=45)
plt.legend()
plt.grid(True, alpha=0.3)

# Overfitting detection
plt.subplot(2, 3, 5)
overfitting_score = [(results[name]['test_mse'] - results[name]['train_mse']) / results[name]['train_mse'] 
                     for name in model_names]
colors = ['green' if score < 0.5 else 'orange' if score < 1.5 else 'red' for score in overfitting_score]

plt.bar(range(len(model_names)), overfitting_score, color=colors, alpha=0.7)
plt.xlabel('Models')
plt.ylabel('Overfitting Score\n(Test MSE - Train MSE) / Train MSE')
plt.title('Overfitting Detection', fontweight='bold')
plt.xticks(range(len(model_names)), [name.split('(')[0].strip() for name in model_names], rotation=45)
plt.axhline(y=0.5, color='orange', linestyle='--', alpha=0.7, label='Moderate')
plt.axhline(y=1.5, color='red', linestyle='--', alpha=0.7, label='High')
plt.legend()
plt.grid(True, alpha=0.3)

# Learning curves
plt.subplot(2, 3, 6)
from sklearn.model_selection import learning_curve

# Use the overfitting model (degree 10)
train_sizes, train_scores, val_scores = learning_curve(
    models['Polynomial (degree=10)'], X_train, y_train,
    train_sizes=np.linspace(0.1, 1.0, 10), cv=5, scoring='neg_mean_squared_error'
)

plt.plot(train_sizes, -train_scores.mean(axis=1), 'o-', color='blue', label='Training')
plt.fill_between(train_sizes, -train_scores.mean(axis=1) - train_scores.std(axis=1),
                 -train_scores.mean(axis=1) + train_scores.std(axis=1), alpha=0.1, color='blue')
plt.plot(train_sizes, -val_scores.mean(axis=1), 'o-', color='red', label='Validation')
plt.fill_between(train_sizes, -val_scores.mean(axis=1) - val_scores.std(axis=1),
                 -val_scores.mean(axis=1) + val_scores.std(axis=1), alpha=0.1, color='red')
plt.xlabel('Training Set Size')
plt.ylabel('Mean Squared Error')
plt.title('Learning Curve\n(Polynomial degree=10)', fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Print detailed results
print("\n=== DETAILED RESULTS ===")
for name, metrics in results.items():
    print(f"\n{name}:")
    print(f"  Training MSE: {metrics['train_mse']:.4f}")
    print(f"  Test MSE: {metrics['test_mse']:.4f}")
    print(f"  Training R²: {metrics['train_r2']:.4f}")
    print(f"  Test R²: {metrics['test_r2']:.4f}")
    overfitting = (metrics['test_mse'] - metrics['train_mse']) / metrics['train_mse']
    print(f"  Overfitting Score: {overfitting:.4f}")
    
    if overfitting < 0.5:
        status = "Good (No overfitting)"
    elif overfitting < 1.5:
        status = "Moderate overfitting"
    else:
        status = "High overfitting"
    print(f"  Status: {status}")
```

**Penjelasan Teoretis Testing dan Validation:**

**Train-Test Split:**
- **Tujuan:** Mengevaluasi performa model pada data yang tidak pernah dilihat selama training
- **Mengapa penting?** Untuk mendeteksi overfitting dan memperkirakan performa real-world
- **Best practices:** 80-20 atau 70-30 split, pastikan representatif, stratified untuk imbalanced data

**Cross-Validation:**
- **Mengapa lebih baik dari simple train-test split?** Memberikan estimasi yang lebih stabil dan menggunakan semua data
- **Bagaimana cara kerjanya?** Membagi data menjadi K fold, setiap fold sekali menjadi validation set
- **Trade-off:** Lebih akurat tapi lebih computationally expensive

**Hyperparameter Tuning:**
- **Masalah dengan test set:** Jika kita tune hyperparameter berdasarkan test set, maka test set tidak lagi "unseen"
- **Solusi:** Gunakan validation set atau cross-validation untuk tuning, test set hanya untuk final evaluation
- **Holdout validation:** Train set → Validation set → Test set (3-way split)

## 6. Data Mismatch Problem

```python
# Simulasi masalah Data Mismatch
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Simulasi skenario: Mobile app untuk klasifikasi gambar bunga
print("=== SKENARIO: MOBILE FLOWER CLASSIFICATION APP ===")
print("- Training data: Foto bunga dari web (jutaan foto)")
print("- Production data: Foto dari mobile camera")
print("- Problem: Domain mismatch antara web photos vs mobile photos")

# Generate synthetic data untuk simulasi
np.random.seed(42)

# Web photos (clean, high quality, good lighting)
n_web = 1000
web_photos_class0 = np.random.multivariate_normal([2, 2], [[0.5, 0.1], [0.1, 0.5]], n_web//2)
web_photos_class1 = np.random.multivariate_normal([4, 4], [[0.5, -0.1], [-0.1, 0.5]], n_web//2)
web_photos = np.vstack([web_photos_class0, web_photos_class1])
web_labels = np.hstack([np.zeros(n_web//2), np.ones(n_web//2)])

# Mobile photos (noisy, variable quality, different lighting)
n_mobile = 200
mobile_photos_class0 = np.random.multivariate_normal([1.8, 2.2], [[1.0, 0.3], [0.3, 1.0]], n_mobile//2)
mobile_photos_class1 = np.random.multivariate_normal([4.2, 3.8], [[1.0, -0.3], [-0.3, 1.0]], n_mobile//2)
mobile_photos = np.vstack([mobile_photos_class0, mobile_photos_class1])
mobile_labels = np.hstack([np.zeros(n_mobile//2), np.ones(n_mobile//2)])

# Split data sesuai Andrew Ng's recommendation
# Training set: Web photos (except train-dev portion)
# Train-dev set: Portion of web photos
# Validation set: Mobile photos
# Test set: Mobile photos

from sklearn.model_selection import train_test_split

# Split web photos into training and train-dev
web_train, web_traindev, web_train_labels, web_traindev_labels = train_test_split(
    web_photos, web_labels, test_size=0.2, random_state=42, stratify=web_labels)

# Split mobile photos into validation and test
mobile_val, mobile_test, mobile_val_labels, mobile_test_labels = train_test_split(
    mobile_photos, mobile_labels, test_size=0.5, random_state=42, stratify=mobile_labels)

print(f"\nDATA SPLITS:")
print(f"Training set (web): {len(web_train)} samples")
print(f"Train-dev set (web): {len(web_traindev)} samples") 
print(f"Validation set (mobile): {len(mobile_val)} samples")
print(f"Test set (mobile): {len(mobile_test)} samples")

# Train model
model = SVC(kernel='rbf', C=1.0, random_state=42)
model.fit(web_train, web_train_labels)

# Evaluate on different sets
train_accuracy = accuracy_score(web_train_labels, model.predict(web_train))
traindev_accuracy = accuracy_score(web_traindev_labels, model.predict(web_traindev))
val_accuracy = accuracy_score(mobile_val_labels, model.predict(mobile_val))
test_accuracy = accuracy_score(mobile_test_labels, model.predict(mobile_test))

print(f"\nMODEL PERFORMANCE:")
print(f"Training accuracy (web): {train_accuracy:.3f}")
print(f"Train-dev accuracy (web): {traindev_accuracy:.3f}")
print(f"Validation accuracy (mobile): {val_accuracy:.3f}")
print(f"Test accuracy (mobile): {test_accuracy:.3f}")

# Diagnosis
print(f"\n=== DIAGNOSIS ===")
overfitting = train_accuracy - traindev_accuracy
data_mismatch = traindev_accuracy - val_accuracy

print(f"Overfitting gap: {overfitting:.3f}")
print(f"Data mismatch gap: {data_mismatch:.3f}")

if overfitting > 0.05:
    print("❌ OVERFITTING DETECTED")
    print("Solutions: Regularization, more data, simpler model")
else:
    print("✅ No significant overfitting")

if data_mismatch > 0.1:
    print("❌ DATA MISMATCH DETECTED") 
    print("Solutions: Domain adaptation, data augmentation, collect more mobile data")
else:
    print("✅ No significant data mismatch")

# Visualisasi
plt.figure(figsize=(15, 10))

# Data distribution
plt.subplot(2, 3, 1)
plt.scatter(web_photos[:, 0], web_photos[:, 1], c=web_labels, alpha=0.6, 
           cmap='viridis', s=20, label='Web Photos')
plt.title('Web Photos Distribution', fontweight='bold')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar()
plt.grid(True, alpha=0.3)

plt.subplot(2, 3, 2)
plt.scatter(mobile_photos[:, 0], mobile_photos[:, 1], c=mobile_labels, alpha=0.6, 
           cmap='viridis', s=30, label='Mobile Photos')
plt.title('Mobile Photos Distribution', fontweight='bold')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar()
plt.grid(True, alpha=0.3)

# Combined view
plt.subplot(2, 3, 3)
plt.scatter(web_photos[:, 0], web_photos[:, 1], c=web_labels, alpha=0.4, 
           cmap='viridis', s=10, label='Web', marker='o')
plt.scatter(mobile_photos[:, 0], mobile_photos[:, 1], c=mobile_labels, alpha=0.8, 
           cmap='viridis', s=40, label='Mobile', marker='^')
plt.title('Data Mismatch Visualization', fontweight='bold')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True, alpha=0.3)

# Performance comparison
plt.subplot(2, 3, 4)
datasets = ['Training\n(Web)', 'Train-dev\n(Web)', 'Validation\n(Mobile)', 'Test\n(Mobile)']
accuracies = [train_accuracy, traindev_accuracy, val_accuracy, test_accuracy]
colors = ['blue', 'green', 'orange', 'red']

bars = plt.bar(datasets, accuracies, color=colors, alpha=0.7)
plt.ylabel('Accuracy')
plt.title('Performance Across Datasets', fontweight='bold')
plt.ylim(0, 1)

# Add value labels on bars
for bar, acc in zip(bars, accuracies):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, 
             f'{acc:.3f}', ha='center', va='bottom')

plt.grid(True, alpha=0.3)

# Error analysis
plt.subplot(2, 3, 5)
error_types = ['Overfitting', 'Data Mismatch']
error_magnitudes = [max(0, overfitting), max(0, data_mismatch)]
colors = ['red' if err > 0.05 else 'green' for err in error_magnitudes]

plt.bar(error_types, error_magnitudes, color=colors, alpha=0.7)
plt.ylabel('Error Gap')
plt.title('Error Analysis', fontweight='bold')
plt.axhline(y=0.05, color='orange', linestyle='--', alpha=0.7, label='Threshold')
plt.legend()
plt.grid(True, alpha=0.3)

# Solutions flowchart (text-based)
plt.subplot(2, 3, 6)
plt.text(0.5, 0.9, 'Data Mismatch Solutions', ha='center', va='top', 
         fontsize=14, fontweight='bold', transform=plt.gca().transAxes)

solutions = [
    '1. Domain Adaptation:',
    '  - Transfer learning',
    '  - Fine-tuning on mobile data',
    '',
    '2. Data Augmentation:',
    '  - Simulate mobile conditions',
    '  - Add noise, blur, lighting changes',
    '',
    '3. Collect More Mobile Data:',
    '  - Representative sampling',
    '  - Active learning'
]

y_pos = 0.75
for solution in solutions:
    plt.text(0.05, y_pos, solution, ha='left', va='top', 
             transform=plt.gca().transAxes, fontsize=10)
    y_pos -= 0.08

plt.xlim(0, 1)
plt.ylim(0, 1)
plt.axis('off')

plt.tight_layout()
plt.show()

print("\n=== DATA MISMATCH SOLUTIONS ===")
print("\n1. DOMAIN ADAPTATION:")
print("   - Preprocessing: Membuat web photos mirip mobile photos")
print("   - Transfer learning: Pre-train on web, fine-tune on mobile")
print("   - Domain adversarial training")

print("\n2. DATA AUGMENTATION:")
print("   - Menambah noise pada web photos")
print("   - Simulasi kondisi mobile (blur, lighting, compression)")
print("   - Geometric transformations")

print("\n3. COLLECT MORE REPRESENTATIVE DATA:")
print("   - Active learning untuk memilih sampel mobile yang informatif")
print("   - Crowdsourcing dengan panduan yang jelas")
print("   - Synthetic data generation")

print("\n=== ANDREW NG'S DIAGNOSTIC FRAMEWORK ===")
print("1. Train-dev set membantu membedakan overfitting vs data mismatch")
print("2. Jika train accuracy >> train-dev accuracy → Overfitting")
print("3. Jika train-dev accuracy >> validation accuracy → Data mismatch")
print("4. Kedua masalah bisa terjadi bersamaan")
print("5. Solusi berbeda untuk setiap masalah")
```

**Penjelasan Teoretis Data Mismatch:**

**Apa itu Data Mismatch?**
- **Definisi:** Perbedaan distribusi antara data training dan data production
- **Mengapa terjadi?** Training data mudah didapat tapi tidak representatif terhadap use case sesungguhnya
- **Dampak:** Model performa bagus di training tapi buruk di production

**Andrew Ng's Diagnostic Framework:**
- **Train-dev set:** Subset dari training data yang tidak digunakan untuk training
- **Tujuan:** Membedakan apakah masalah karena overfitting atau data mismatch
- **Diagnosis:** 
  - High training error → Underfitting
  - Training error << Train-dev error → Overfitting  
  - Train-dev error << Validation error → Data mismatch

**Solusi Data Mismatch:**
- **Domain adaptation:** Mengadaptasi model untuk domain yang berbeda
- **Data preprocessing:** Membuat training data lebih mirip production data
- **Data collection:** Mengumpulkan lebih banyak data yang representatif

## 7. No Free Lunch Theorem

```python
# Demonstrasi No Free Lunch Theorem
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

print("=== NO FREE LUNCH THEOREM ===")
print("Teorema yang menyatakan bahwa:")
print("- Tidak ada algoritma yang universal terbaik untuk semua masalah")
print("- Performa algoritma sangat bergantung pada karakteristik data")
print("- Yang terbaik pada satu dataset bisa terburuk pada dataset lain")

# Generate different types of datasets
np.random.seed(42)
n_samples = 200

def generate_datasets():
    """Generate datasets dengan karakteristik berbeda"""
    datasets = {}
    
    # 1. Linear relationship
    X1 = np.random.uniform(-2, 2, n_samples).reshape(-1, 1)
    y1 = 3 * X1.ravel() + 2 + np.random.normal(0, 0.3, n_samples)
    datasets['Linear'] = (X1, y1)
    
    # 2. Non-linear (polynomial)
    X2 = np.random.uniform(-2, 2, n_samples).reshape(-1, 1)
    y2 = X2.ravel()**3 - 2*X2.ravel()**2 + X2.ravel() + np.random.normal(0, 0.5, n_samples)
    datasets['Polynomial'] = (X2, y2)
    
    # 3. Step function (tree-like)
    X3 = np.random.uniform(-2, 2, n_samples).reshape(-1, 1)
    y3 = np.where(X3.ravel() < -1, -2, 
         np.where(X3.ravel() < 0, 0,
         np.where(X3.ravel() < 1, 2, 4))) + np.random.normal(0, 0.2, n_samples)
    datasets['Step Function'] = (X3, y3)
    
    # 4. Noisy with outliers
    X4 = np.random.uniform(-2, 2, n_samples).reshape(-1, 1)
    y4 = 2 * X4.ravel() + 1 + np.random.normal(0, 1, n_samples)
    # Add outliers
    outlier_indices = np.random.choice(n_samples, 20, replace=False)
    y4[outlier_indices] += np.random.normal(0, 5, 20)
    datasets['Noisy with Outliers'] = (X4, y4)
    
    return datasets

# Initialize models
models = {
    'Linear Regression': LinearRegression(),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42),
    'SVR (RBF)': SVR(kernel='rbf', C=1.0, gamma='scale'),
    'K-Nearest Neighbors': KNeighborsRegressor(n_neighbors=5)
}

# Generate datasets
datasets = generate_datasets()

# Test all models on all datasets
results = {}
plt.figure(figsize=(20, 15))

for i, (dataset_name, (X, y)) in enumerate(datasets.items()):
    results[dataset_name] = {}
    
    # Split data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    
    # Test each model
    for j, (model_name, model) in enumerate(models.items()):
        # Train and predict
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        
        # Calculate error
        mse = mean_squared_error(y_test, y_pred)
        results[dataset_name][model_name] = mse
        
        # Plot results
        subplot_idx = i * len(models) + j + 1
        plt.subplot(len(datasets), len(models), subplot_idx)
        
        # Plot data
        plt.scatter(X_test, y_test, alpha=0.6, s=20, color='blue', label='Test Data')
        
        # Plot predictions
        X_plot = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
        try:
            y_plot = model.predict(X_plot)
            plt.plot(X_plot, y_plot, 'r-', linewidth=2, alpha=0.8, label='Prediction')
        except:
            pass  # Some models might fail on certain data
        
        plt.title(f'{dataset_name}\n{model_name}\nMSE: {mse:.3f}', fontsize=10)
        plt.xlabel('X')
        plt.ylabel('y')
        if j == 0:  # Only show legend for first column
            plt.legend(fontsize=8)
        plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Performance comparison
print("\n=== PERFORMANCE COMPARISON ===")
print("Model Performance (MSE) on Different Datasets:")
print("-" * 80)

# Create performance matrix
performance_matrix = []
model_names = list(models.keys())
dataset_names = list(datasets.keys())

print(f"{'Dataset':<20}", end="")
for model_name in model_names:
    print(f"{model_name[:15]:<15}", end="")
print()
print("-" * 80)

for dataset_name in dataset_names:
    print(f"{dataset_name:<20}", end="")
    row = []
    for model_name in model_names:
        mse = results[dataset_name][model_name]
        row.append(mse)
        print(f"{mse:<15.3f}", end="")
    performance_matrix.append(row)
    print()

# Find best model for each dataset
print("\n=== BEST MODEL FOR EACH DATASET ===")
for i, dataset_name in enumerate(dataset_names):
    best_idx = np.argmin(performance_matrix[i])
    best_model = model_names[best_idx]
    best_score = performance_matrix[i][best_idx]
    print(f"{dataset_name:<20}: {best_model} (MSE: {best_score:.3f})")

# Rank models across all datasets
print("\n=== AVERAGE RANKING ACROSS ALL DATASETS ===")
performance_matrix = np.array(performance_matrix)
rankings = []

for i in range(len(dataset_names)):
    # Get ranking for this dataset (1 = best, 4 = worst)
    dataset_ranking = np.argsort(performance_matrix[i]) + 1
    rankings.append(dataset_ranking)

rankings = np.array(rankings)
avg_rankings = rankings.mean(axis=0)

print("Average Rank (1=best, 4=worst):")
for i, model_name in enumerate(model_names):
    print(f"{model_name:<25}: {avg_rankings[i]:.2f}")

# Visualize rankings
plt.figure(figsize=(12, 8))

# Heatmap of MSE values
plt.subplot(2, 2, 1)
im = plt.imshow(performance_matrix, cmap='RdYlBu_r', aspect='auto')
plt.colorbar(im, label='MSE')
plt.xticks(range(len(model_names)), [name[:10] for name in model_names], rotation=45)
plt.yticks(range(len(dataset_names)), dataset_names)
plt.title('Performance Heatmap\n(Darker = Better)', fontweight='bold')

# Rankings heatmap
plt.subplot(2, 2, 2)
im2 = plt.imshow(rankings, cmap='RdYlGn_r', aspect='auto')
plt.colorbar(im2, label='Rank')
plt.xticks(range(len(model_names)), [name[:10] for name in model_names], rotation=45)
plt.yticks(range(len(dataset_names)), dataset_names)
plt.title('Ranking Heatmap\n(Green = Better Rank)', fontweight='bold')

# Average rankings bar plot
plt.subplot(2, 2, 3)
colors = ['green' if rank < 2.5 else 'orange' if rank < 3.5 else 'red' for rank in avg_rankings]
bars = plt.bar(range(len(model_names)), avg_rankings, color=colors, alpha=0.7)
plt.xticks(range(len(model_names)), [name[:10] for name in model_names], rotation=45)
plt.ylabel('Average Rank')
plt.title('Average Ranking Across All Datasets', fontweight='bold')
plt.axhline(y=2.5, color='red', linestyle='--', alpha=0.5, label='Threshold')
plt.legend()
plt.grid(True, alpha=0.3)

# Wins count
plt.subplot(2, 2, 4)
wins = [np.sum(rankings[:, i] == 1) for i in range(len(model_names))]
plt.bar(range(len(model_names)), wins, alpha=0.7, color='skyblue')
plt.xticks(range(len(model_names)), [name[:10] for name in model_names], rotation=45)
plt.ylabel('Number of Wins')
plt.title('Number of Times Each Model\nWas Best', fontweight='bold')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\n=== NO FREE LUNCH THEOREM INSIGHTS ===")
print("Dari eksperimen di atas, kita bisa melihat bahwa:")
print(f"1. Tidak ada model yang selalu terbaik di semua dataset")
print(f"2. Model terbaik bergantung pada karakteristik data:")

# Analyze which model works best for which type of data
print("\n   Dataset Analysis:")
for i, dataset_name in enumerate(dataset_names):
    best_idx = np.argmin(performance_matrix[i])
    best_model = model_names[best_idx]
    
    if 'Linear' in dataset_name:
        print(f"   - {dataset_name}: {best_model} (sesuai ekspektasi untuk data linear)")
    elif 'Polynomial' in dataset_name:
        print(f"   - {dataset_name}: {best_model} (baik untuk non-linear complex)")
    elif 'Step' in dataset_name:
        print(f"   - {dataset_name}: {best_model} (bagus untuk data diskrit/categorical)")
    elif 'Noisy' in dataset_name:
        print(f"   - {dataset_name}: {best_model} (robust terhadap outliers)")

print(f"\n3. Implikasi praktis:")
print(f"   - Selalu test multiple algorithms")
print(f"   - Understand your data characteristics")
print(f"   - Use domain knowledge untuk memilih model")
print(f"   - Cross-validation penting untuk model selection")

print(f"\n=== PRACTICAL GUIDELINES ===")
print("Berdasarkan No Free Lunch Theorem:")
print("1. JANGAN langsung memilih satu algoritma 'favorit'")
print("2. SELALU coba beberapa algoritma berbeda")
print("3. GUNAKAN cross-validation untuk perbandingan yang fair")
print("4. PERTIMBANGKAN karakteristik data dan domain problem")
print("5. KOMBINASIKAN domain knowledge dengan empirical testing")
```

**Penjelasan Teoretis No Free Lunch Theorem:**

**Konsep Dasar:**
- **Teorema:** Tidak ada algoritma pembelajaran yang superior untuk semua kemungkinan masalah
- **Implikasi:** Performance rata-rata semua algoritma sama jika dihitung untuk semua kemungkinan problem
- **Praktis:** Pada problem spesifik, beberapa algoritma akan lebih baik dari yang lain

**Mengapa Teorema Ini Penting:**
- **Menghindari "Silver Bullet" Mentality:** Tidak ada satu algoritma yang bisa menyelesaikan semua masalah
- **Pentingnya Empirical Testing:** Harus test multiple algoritma pada data spesifik
- **Domain Knowledge Matters:** Memahami karakteristik data membantu pemilihan algoritma

**Implikasi Praktis:**
- **Model Selection:** Selalu compare multiple algorithms
- **No Assumptions:** Jangan assume algoritma yang bagus di satu domain akan bagus di domain lain
- **Data-Driven:** Decision harus berdasarkan evidence dari data, bukan bias personal
